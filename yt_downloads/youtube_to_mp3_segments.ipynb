```python
#imports

from pytube import YouTube
from pytube import Playlist
import pydub
import audiosegment
from pydub import AudioSegment
import os
```

```python
#get all urls from playlists 

def get_playlist(playlist_links): 
    urls = []
    for playlist in playlist_links:
        playlist_urls = Playlist(playlist)
        for url in playlist_urls: 
            urls.append(url)
    return urls
```

```python
#save as mp3s (okay i know they're mp4s but it's just audio i promise)
save_path = "mp3s"
num = 0

def download_mp3(url, num): 
    print("Handling file number" + str(num))
    yt = YouTube(url)
    stream = yt.streams.filter(only_audio=True).first()
    out_file = stream.download(output_path=save_path)
    base, ext = os.path.splitext(out_file)
    new_file = base + '.mp3'
    os.rename(out_file, new_file)
    print("Done with file no " + str(num))
    num += 1
    return num
```

```python
#segmenting 
file_path = "segments/"
def segment_mp3(file):
    base = file[5:-3]
    num = 0
    song = audiosegment.AudioSegment(pydub.AudioSegment.from_file(file), "lofi-bytes")
    segments = song.dice(45, zero_pad = True)
    for segment in segments: 
        print("Downloading segment " + str(num) + " from " + base)
        segment.export(file_path + base + str(num) + ".mp3", format="mp3")
        num += 1
```

```python
directory = 'mp3s'

def segment_download():
    for filename in os.listdir(directory): 
        f = os.path.join(directory, filename)
        if os.path.isfile(f):
            segment_mp3(f)
```

```python
def main(): 
    num = 0
    playlist_links = ["https://www.youtube.com/playlist?list=PLRA7V6h5tgpD3E1Fei8PyKW6tmtkzKzFR"]
    urls = get_playlist(playlist_links)
    for url in urls: 
        num = download_mp3(url, num)
    segment_download()

main()
```
